```python
# function to get sub direcotories name and subdirectories name with path

def get_subdir_name_n_path(src_path):
  sub_dir = []
  sub_dir_w_path = []
  # getting sub-directories
  for (dir_path, dir_names, file_names) in walk(src_path):
    sub_dir.extend(dir_names)
    break
  # assigning sub-directories name
  for i in range(len(sub_dir)):
    sub_path = os.path.join(src_path+'/'+sub_dir[i]+'/')
    sub_dir_w_path.append(sub_path)

  return sub_dir, sub_dir_w_path
```

```python
# checking the function with our root directory
fold_names, fold_paths = get_subdir_name_n_path(dir_path)

# fold_names are resume classes, and fold_paths is path of that class
print(f"Total class or folder in the dataset:\n{fold_names}\n")
print(f"Path of those folders:\n{fold_paths}")
```

```python
# function to get files within a sub-directory

def get_subdir_files(src_dir):
  sub_files = []
  only_names = []
  for file in os.listdir(src_dir):
    if(os.path.isfile(os.path.join(src_dir, file))):
      file_name = os.path.join(src_dir, file)
      sub_files.append(file_name)
      only_names.append(file)
  return sub_files, only_names
```

```python
# function to get all files within subdirectoris as well
def get_all_files(src_dir):
  all_files = []
  for path, subdirs, files in os.walk(src_dir):
    for name in files:
      # file name with path
      file_name = os.path.join(path, name)
      all_files.append(file_name)
  return all_files
```

```python
# functon to create a directory
def make_directory(root_path, dir_name):
  # defining path
  path = os.path.join(root_path, dir_name)

  # we can give a directory name that already exist by mistake, therefore, using try except
  try:
    os.mkdir(path)
    print(f"Directory created successfully at: {path}")
  except OSError as error:
      print(error)
  return path
```

```python
# function to move a file from one folder to another folder
import shutil

def move_file(src_path, dest_path):
  shutil.move(src_path, dest_path)
  print(src_path, " moved...")
```

```python
# as we are having 2494 resumes for all classes, and we supposed to split the dataset into train, test, and validation
# i will make 10% data from each class as test set which will be totally unseen,
# and rest of the data will be splitted into train and validation during the training time
# test data will be moved into our newly created 'test' folder

# first of all, I will create 24 sub-directory for 24 classes according to their name
for i in range(len(fold_names)):
  # creating a folder in the test directory for the class
  f_path = make_directory(test_path+'/', fold_names[i])

  # getting the corresponding files of current sub-directory
  sub_resumes_path, sub_resumes_name = get_subdir_files(fold_paths[i])

  # now define the number of files to move (10%)
  file_to_move = round((len(sub_resumes_path) * 10)/ 100)
  print(f"Resumes to be moved for {fold_names[i]}: {file_to_move}")

  # moving files from each train class to test class
  cnt = 0
  for k in range(len(sub_resumes_path)):
    if(cnt > file_to_move):
      break
    else:
      move_file(sub_resumes_path[k], f_path+'/'+sub_resumes_name[k])
      cnt += 1
  print(f"Total {cnt} resumes moved for {fold_names[i]} !")
```

```python
# function to convert all characters in lower case and remove unnecessay characters and words
import re
import nltk

def preprocess_text(txt):
    # convert all characters in the string to lower case
    txt = txt.lower()
    # remove non-english characters, punctuation and numbers
    txt = re.sub('[^a-zA-Z]', ' ', txt)
    txt = re.sub('http\S+\s*', ' ', txt)  # remove URLs
    txt = re.sub('RT|cc', ' ', txt)  # remove RT and cc
    txt = re.sub('#\S+', '', txt)  # remove hashtags
    txt = re.sub('@\S+', '  ', txt)  # remove mentions
    txt = re.sub('\s+', ' ', txt)  # remove extra whitespace
    # tokenize word
    txt = nltk.tokenize.word_tokenize(txt)
    # remove stop words
    txt = [w for w in txt if not w in nltk.corpus.stopwords.words('english')]

    return ' '.join(txt)
```

```python
# function to extract text from a PDF file
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, "rb") as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        text = ""
        for page in pdf_reader.pages:
            text += page.extract_text()
    return text
```
